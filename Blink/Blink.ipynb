```python
from pynq import Overlay
from pynq import MMIO
import time

# Load overlay 
ol = Overlay('Blink_wrapper.bit')

# BRAM base address
BRAM_BASE = 0x40000000

# Map the 4 KB BRAM address space starting at BRAM_BASE for MMIO access
bram = MMIO(BRAM_BASE, 0x1000)

while True:
    # Write 1 to BRAM — LED ON
    bram.write(0, 1)
    print("LED ON")
    time.sleep(1)

    # Write 0 to BRAM — LED OFF
    bram.write(0, 0)
    print("LED OFF")
    time.sleep(1)
```
